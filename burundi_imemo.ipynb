```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Define the data arrays
data = {'Индекс ЭВВ Бурунди': [8.5, 8.5, 8.5, 8.5, 8.5, 8, 7, 8, 5.5, 7.5, 7.5, 7.5, 6.5, 6, 9, 9, 8.5, 8, 7.5, 6.5, 7],
        'Политическая стабильность Бурунди': [-2.15, -2.28, -2.24, -2.52, -1.52, -1.43, -1.38, -1.64, -1.27, -1.63, -1.81, -1.70, -1.35, -0.78, -1.94, -1.97, -2.01, -1.57, -1.62, -1.54, -1.36]}
df = pd.DataFrame(data)

# Add a constant column to the data
X = sm.add_constant(df['Индекс ЭВВ Бурунди'])

# Create a linear regression model and fit the data
model = sm.OLS(df['Политическая стабильность Бурунди'], X).fit()

# Print the summary statistics of the model
print(model.summary())

# Calculate the slope and intercept of the trend line
x = df['Индекс ЭВВ Бурунди']
y = df['Политическая стабильность Бурунди']
slope, intercept = np.polyfit(x, y, 1)

# Create a scatter plot with the trend line
plt.scatter(x, y)
plt.plot(x, slope*x + intercept, color='blue')

# Add axis labels and a title
plt.xlabel('Индекс ЭВВ Бурунди')
plt.ylabel('Политическая стабильность Бурунди')
plt.title('Зависимость политической стабильности от Индекса ЭВВ в Бурунди')

# Show the plot
plt.show()
```
